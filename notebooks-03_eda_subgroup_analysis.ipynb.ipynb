### VISUALIZATION OF THE FREQUENCY TABLES

```python

# Get county frequencies
county_counts = df["County"].value_counts()

# Plot
plt.figure(figsize=(12,6))
county_counts.plot(kind="bar")

plt.title("County Distribution of Respondents")
plt.xlabel("County")
plt.ylabel("Number of Respondents")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The county distribution shows variation in respondent representation across counties. Some counties have significantly higher respondent counts, indicating uneven sampling distribution across regions.

```python
# Gender frequency (A10i = Gender)
gender_counts = df["A10i"].value_counts()

# Plot
plt.figure(figsize=(6,4))
gender_counts.plot(kind="bar")

plt.title("Gender Distribution of Respondents")
plt.xlabel("Gender")
plt.ylabel("Number of Respondents")

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

The gender distribution shows the proportion of male and female respondents in the dataset. This helps assess whether the sample is balanced or skewed toward a particular gender group.

- This helps evaluate sampling bias
- Useful for comparing with other variables (like education or county)

```python
# Rural/Urban frequency (A9)
residence_counts = df["A9"].value_counts()

# Plot
plt.figure(figsize=(6,4))
residence_counts.plot(kind="bar")

plt.title("Rural vs Urban Distribution of Respondents")
plt.xlabel("Location Type")
plt.ylabel("Number of Respondents")

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

The rural vs urban distribution shows how respondents are split between rural and urban areas. This helps in understanding the geographical representation of the dataset and whether the sample is biased toward rural or urban populations.

THis helps plot helps us the following:
- Whether your dataset is mostly rural or urban
- If there is geographic imbalance
- Useful for comparing with:
- - education (A21)
- - income/financial access variables
- - county distribution

```python
education_counts = df["A21"].value_counts()

# Plot
plt.figure(figsize=(10,5))
education_counts.plot(kind="bar")

plt.title("Education Level Distribution")
plt.xlabel("Education Level")
plt.ylabel("Number of Respondents")

plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

The education distribution shows the highest level of education attained by respondents. The majority of respondents are concentrated in primary and secondary education levels, while a smaller proportion have attained university education. This indicates relatively low higher education attainment within the sample population.

```python
age_counts = df["A14v"].value_counts().sort_index()

# Plot
plt.figure(figsize=(8,5))
age_counts.plot(kind="bar")

plt.title("Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Number of Respondents")

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

The age group distribution shows that the majority of respondents fall within the lower coded age categories, indicating a relatively younger population in the dataset. The number of respondents decreases as the age categories increase, showing fewer older individuals in the sample

#### Exclusion rate by county

```python
df['excluded_informal_new2022'].unique()
```

```python
df['financial_exclusion'] = df['excluded_informal_new2022'].map({'Excluded': 1,'Formal': 0,'mobileonly': 0})
```

```python
county_exclusion = (df.groupby('County')['financial_exclusion'].mean().sort_values(ascending=False)) * 100

print(county_exclusion)
```

```python
plt.figure(figsize=(16,8))

county_exclusion.plot(kind='bar')

plt.title('Financial Exclusion Rate by County')
plt.xlabel('County')
plt.ylabel('Exclusion Rate (%)')

plt.xticks(rotation=90)

plt.show()
```

```python
rural_urban_exclusion = (df.groupby('A9')['financial_exclusion'].mean()) * 100

print(rural_urban_exclusion)
```

```python
plt.figure(figsize=(8,6))

rural_urban_exclusion.plot(kind='bar')

plt.title('Financial Exclusion Rate: Rural vs Urban')
plt.xlabel('Residence')
plt.ylabel('Exclusion Rate (%)')

plt.xticks(rotation=0)

plt.show()
```

```python
top10_highest_exclusion = county_exclusion.head(10)

print(top10_highest_exclusion)
```

```python
plt.figure(figsize=(12,6))

top10_highest_exclusion.plot(kind='bar')

plt.title('Top 10 Counties with Highest Financial Exclusion')
plt.xlabel('County')
plt.ylabel('Exclusion Rate (%)')

plt.xticks(rotation=45)

plt.show()
```

```python
top10_lowest_exclusion = county_exclusion.tail(10)

print(top10_lowest_exclusion)
```

```python
plt.figure(figsize=(12,6))

top10_lowest_exclusion.plot(kind='bar')

plt.title('Top 10 Counties with Lowest Financial Exclusion')
plt.xlabel('County')
plt.ylabel('Exclusion Rate (%)')

plt.xticks(rotation=45)

plt.show()
```
