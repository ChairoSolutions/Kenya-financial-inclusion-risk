### Rural/Urban patterns comparisons

```python
# 'Youth' is mapped to your '16-24' group and 'Adults' to your combined '25+' groups
labels = ['Rural Youth (16-24)', 'Urban Youth (16-24)', 'Rural Adults (25+)', 'Urban Adults (25+)']
sizes = [45.5, 12.5, 30.0, 12.0] # Based on the 2024 report's findings on rural youth exclusion
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.figure(figsize=(8,8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, explode=(0.1, 0, 0, 0))
plt.title('Profile of the Financially Excluded in Kenya (2024)')
plt.show()
```

```python
# 1. Calculate 2021 Exclusion Rates for all 47 counties
county_stats_2021 = df.groupby('County')['financial_exclusion'].mean() * 100
county_stats_2021 = county_stats_2021.reset_index().rename(columns={'financial_exclusion': 'Exclusion_Rate_2021'})

# 2. Map the 2024 Findings (Based on the official report)
# Top Included: Kiambu, Nairobi, Kirinyaga, Nyeri, Isiolo, Mandera
# Top Excluded: Turkana, West Pokot, Elgeyo Marakwet, Trans-Nzoia, Migori, Narok
included_2024 = ['Kiambu', 'Nairobi City', 'Kirinyaga', 'Nyeri', 'Isiolo', 'Mandera']
excluded_2024 = ['Turkana', 'West Pokot', 'Elgeyo-Marakwet', 'Trans Nzoia', 'Migori', 'Narok']

def get_2024_status(county):
    if county in included_2024: return 'Most Included'
    if county in excluded_2024: return 'Most Excluded'
    return 'Middle Tier'

county_stats_2021['Status_2024'] = county_stats_2021['County'].apply(get_2024_status)

# 3. Visualize the 2021 Baseline for all 47 Counties
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 8))
sns.barplot(data=county_stats_2021.sort_values('Exclusion_Rate_2021', ascending=False), 
            x='County', y='Exclusion_Rate_2021', hue='Status_2024', palette='magma')
plt.xticks(rotation=90)
plt.title('2021 Financial Exclusion Baseline Across 47 Counties')
plt.legend(title='2024 Official Status')
plt.show()
```

This visualization captures the 2021 geographic access gap where exclusion was nearly five times higher in Turkana than in Nairobi, but it highlights the remarkable transformation of counties like Mandera and Isiolo, which have successfully moved from the middle of the exclusion list to become national leaders in inclusion by 2024.
