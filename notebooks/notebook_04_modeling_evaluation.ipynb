## Why This Is a Classification Problem
This project is a classification problem because the model is predicting a category or class rather than a continuous numerical value.The target variable is:
1 = Financially Excluded
0 = Financially Included. 
Since the model must decide whether an individual belongs to one of these two categories, the task is specifically a binary classification problem.
The goal is not to predict an amount such as income or savings, but to classify individuals into exclusion-risk groups based on demographic, geographic, and financial characteristics.

### Logistic Regression
Logistic Regression is a supervised machine learning algorithm commonly used for binary classification problems. It predicts the probability that an observation belongs to a particular class.
In this project, Logistic Regression can estimate the probability that a Kenyan adult is financially excluded.
#### Why Logistic Regression is Important
* Easy to interpret
* Fast to train

Works well as a baseline model
Provides probability outputs
Commonly used in financial risk analysis
#### Why It Fits This Project
The project needs an interpretable model that can show how factors such as education, employment, rural location, or mobile money usage affect financial exclusion risk.

## Random Forest
Random Forest is an ensemble learning algorithm that combines many decision trees to improve prediction accuracy and reduce overfitting.
Instead of relying on one tree, the model creates multiple trees and combines their predictions.
#### Why Random Forest is Useful
* Handles complex relationships well
* Works with categorical and numerical data
* Reduces overfitting compared to a single decision tree
* Captures nonlinear patterns
* Provides feature importance scores
#### Why It Fits This Project
Financial exclusion is influenced by many interacting socioeconomic factors. Random Forest can capture hidden relationships between variables such as county, education, employment status, and digital finance access.

## Gradient Boosting
Gradient Boosting is another ensemble machine learning method that builds models sequentially. Each new model attempts to correct the errors made by the previous model.
The algorithm gradually improves prediction performance by focusing on difficult cases.
#### Why Gradient Boosting is Useful
* Often achieves high predictive accuracy
* Handles complex nonlinear patterns
* Works well on structured/tabular data
* Can identify subtle risk patterns
#### Why It Fits This Project
Some financially excluded individuals may have patterns that are difficult to detect using simpler models. Gradient Boosting can improve performance by learning these more complex relationships.

## Why Recall Matters
Recall measures how well the model identifies actual positive cases.
In this project, the positive class is:
Financially Excluded Individuals
#### Why Recall Is Important 
The main goal of the project is to identify people who are truly financially excluded.
A low recall score means the model is failing to identify many excluded individuals. These missed cases are called false negatives.
In real-world use
Banks may fail to reach underserved populations
Policymakers may underestimate exclusion levels
Financial inclusion programs may miss vulnerable groups
Because of this, recall is especially important for ensuring vulnerable individuals are correctly identified.

## Planned Evaluation Metrics
The project will evaluate models using:
* Accuracy → Overall correctness of predictions
* Precision → How many predicted excluded individuals are truly excluded
* Recall → How many truly excluded individuals are correctly identified
* F1-Score → Balance between precision and recall
* ROC-AUC → Ability to separate excluded vs included individuals
* Confusion Matrix → Breakdown of prediction performance

## Baseline Model Choice
The baseline model for this project will be Logistic Regression.
#### Why Logistic Regression Was Chosen
* Simple and interpretable
* Fast to implement
* Common baseline for classification tasks
* Produces probability scores
* Easy to compare against more advanced models

Using Logistic Regression as the baseline allows later models such as Random Forest and Gradient Boosting to be evaluated against a simple, understandable starting point.
